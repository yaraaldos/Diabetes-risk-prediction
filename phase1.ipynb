# **This project analyzes  Diabetes Risk Prediction , The dataset contains attributes such as age, gender, and medical symptoms (e.g.,bmi,cr) that  will be used to:**

 *1-Classify patients as diabetic or non-diabetic or potential infection.*

 *2-Cluster patients with similar profiles to discover hidden patterns.*

 *3-Predict and evaluate outcomes in order to provide meaningful insights that support early detection of diabetes.*


# **Dataset Source**

*We used the Diabetes Dataset (Kaggle, by AravindPCoder).*

  

*Link: [https://www.kaggle.com/datasets/aravindpcoder/diabetes-dataset](https://www.kaggle.com/datasets/aravindpcoder/diabetes-dataset)*


 # **The dataset contains 14 attributes including the class label (CLASS):**

1- ID → Numeric (Integer)  → e.g., 19775  
2- No_Pation → Numeric (Integer) → e.g., 34221  
3- Gender → Categorical (Binary: M/F)  
4- AGE → Numeric (Integer) → e.g., 31, 50  
5- Urea → Numeric (Float) → e.g., 4.9  
6- Cr → Numeric (Integer) → e.g., 1.1  
7- HbA1c → Numeric (Float) → e.g., 4.2  
8- Chol → Numeric (Float) → e.g., 4.2  
9- TG → Numeric (Float) → e.g., 1.4  
10- HDL → Numeric (Float) → e.g., 1.4  
11- LDL → Numeric (Float) → e.g., 1.3  
12- VLDL → Numeric (Float) → e.g., 0.4  
13- BMI → Numeric (float) → e.g., 24  
14- CLASS → Nominal (P/Y/N)
